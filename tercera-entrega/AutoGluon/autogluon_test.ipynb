```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split

# Cargar datos
data = TabularDataset('buckets/b1/datasets/competencia_03_ternaria.csv.gz')

# Dividir datos en conjuntos de entrenamiento y prueba
train_data = data[data['foto_mes'] <= 202107]
test_data = data[data['foto_mes'] == 202109]

# Definir la columna objetivo y entrenar el modelo
label_column = 'clase_ternaria'
predictor = TabularPredictor(label=label_column).fit(train_data)

# Hacer predicciones en datos de prueba
predictions = predictor.predict(test_data)

# Obtener probabilidades de las predicciones
probabilities = predictor.predict_proba(test_data)

# Agregar las probabilidades al conjunto de datos de prueba
test_data['probabilidad_baja'] = probabilities.iloc[:, 1]

# Guardar las predicciones y probabilidades
predictions.to_csv('buckets/b1/exp/autogluon_1/predictions.csv', index=False)
test_data.to_csv('buckets/b1/exp/autogluon_1/test_data_with_probabilities.csv', index=False)
```
